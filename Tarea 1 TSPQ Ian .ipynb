1. *Ley de gases ideales*
Esta ecuacion relaciona la presion, volumen, cantidad de sustancia y la temperatura mediante la expresion:

PV = nRT

```python
#Datos

P = 1.2   #atm
T = 273   #K
V = 0.7   #L

#Se busca obtener la cantidad de sustancia por lo que se tiene como incognita a n

#No se tiene el valor de R, pero se puede colocar sabiendo que se tienen como unidades atm,K y L
#Opcion 1
R = input ("El valor de R es: ")
R = float (R)

#Opcion 2
 #R = 0.0825  atm*L / mol * K

#Ec.  PV=nRT

n = (P*V / (R*T))

print (f"La cantidad de sustancia es : {n:.4f} mol")
```

2. *Ecuacion de Arrenius (Cinetica Quimica)*
Esta ecuacion describe la dependencia de la cte de velocidad,k, con la T

  k = A*e^(-Ea/(R*T))

```python
#Datos

import math

A = 5*10**7   #s^-1
Ea = 75000    #J*mol^-1
R = 8.314     #J*mol^-1*K^-1
T = 350       #K

#Calcular la velocidad de K

#Ec. k = A*exp**(-Ea / (R*T) )

k = A * math.exp(-Ea / (R*T) )

print (f"La cte de velocidad es : {k:.4f} 1/s")
```

5. *Numero de Reynolds*

Se emplea para caracterizar el regimen de flujo (Laminar, de transicion o turbulento):

Re = Rho * v * D / Mu




```python
#Datos

Rho = 1000    # kg * m^-3
v = 1.2       # m * s^-1
D = 0.05      # m
Mu = 0.001       # Pa*s

#Calcular el valor de Reynolds

#Ec. Re = Rho * v * D / Mu

Re = Rho * v * D / Mu

print (f"El valor de Reynols es : {Re:.0f}")

```
